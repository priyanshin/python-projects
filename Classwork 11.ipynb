```python
import matplotlib.pyplot as plt
```

```python
x1= [1, 2, 3, 4, 5]
y1 = [5, 10, 15, 20, 25]


x2= [1, 2, 3, 4, 5]
y2= [2, 4, 8, 4, 2]

plt.plot(x1, y1, label= "first line")
plt.plot(x2, y2, label= "second line")


plt.bar(x1, y1, label = "bar", color = "purple")
plt.xlabel("my x value")
plt.ylabel("my y value")
plt.legend()

plt.show()
```

```python
alice_content = open('AliceInWonderland.txt')
```

```python
alice_content.seek(0)

word_counts = {}
word_lengths = []

for line in alice_content: 
    
    words = line.split()
    
    for word in words:
        length = len(word)

        word_lengths.append(length)

        if word_counts.get(length) == None:
            word_counts[length] = 1
        else:
            word_counts[length] += 1
```

```python
print word_counts
```

```python
keys = []
values = []
for key in word_counts:
    keys.append(key)
    values.append(word_counts[key])

plt.bar(keys[1:20], values[1:20])
plt.show()
```
