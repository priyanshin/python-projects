# Document Classification

### Data

I am using sample Amazon Phone reviews data to build Text Classifiers

[Link to download the data](https://data.world/promptcloud/amazon-mobile-phone-reviews)

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# Read in the data
df = pd.read_csv('Amazon_Unlocked_Mobile.csv')

# Sample the data to speed up computation
df = df.sample(frac=0.1, random_state=10)

# View the top rows of the data

df.head()
```

```python
# Check the dimensions of the dataframe
df.shape 
```

```python
# A sample text review
list(df['Reviews'][:2])
```

```python
# Drop missing values
df.dropna(inplace=True)

# Remove any 'neutral' ratings equal to 3
df = df[df['Rating'] != 3]

# Encode 4s and 5s as 1 (rated positively)
# Encode 1s and 2s as 0 (rated poorly)
df['Positively Rated'] = np.where(df['Rating'] > 3, 1, 0)
df.head(5)
```

```python
# Most ratings are positive
df['Positively Rated'].mean()
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Reviews'], 
                                                    df['Positively Rated'], 
                                                    random_state=0)
```

```python
print('X_train first entry:\n\n', X_train.iloc[0])
print('\n\nX_train shape: ', X_train.shape)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Simple Bag of Words Model

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train)
```

```python
vect.get_feature_names()[::2000]
```

```python
len(vect.get_feature_names())
```

```python
# transform the documents in the training data to a document-term matrix
X_train_vectorized = vect.transform(X_train)

X_train_vectorized
```

```python
pd.set_option('display.max_columns', 500)
pd.DataFrame(X_train_vectorized.toarray(), columns=vect.get_feature_names())
```

```python
from sklearn import svm

# Train a SVM model
model = svm.SVC(kernel='linear')
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score

# Predict the transformed test documents
predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
print('Accuracy',accuracy_score(y_test, predictions))
```

```python
# get the feature names as numpy array
feature_names = np.array(vect.get_feature_names())

# Sort the coefficients from the model
sorted_coef_index = model.coef_[0].toarray()[0].argsort()

# Find the 10 smallest and 10 largest coefficients
# The 10 largest coefficients are being indexed using [:-11:-1] 
# so the list returned is in order of largest to smallest
print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

### Tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect = TfidfVectorizer(min_df=5).fit(X_train)
len(vect.get_feature_names())
```

```python
X_train_vectorized = vect.transform(X_train)
pd.DataFrame(X_train_vectorized.toarray(), columns=vect.get_feature_names())
```

```python
model = svm.SVC(kernel='linear')
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
print('Accuracy',accuracy_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names())

sorted_tfidf_index = X_train_vectorized.max(0).toarray()[0].argsort()

print('Smallest tfidf:\n{}\n'.format(feature_names[sorted_tfidf_index[:10]]))
print('Largest tfidf: \n{}'.format(feature_names[sorted_tfidf_index[:-11:-1]]))
```

```python
# These reviews are treated the same by our current model
print(model.predict(vect.transform(['not an issue, phone is working',      # ideally this should be predicted 1
                                    'an issue, phone is not working'])))   # ideally this should be predicted 0
```

### N Grams

```python
# Fit the CountVectorizer to the training data specifiying a minimum 
# document frequency of 5 and extracting 1-grams and 2-grams
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)

X_train_vectorized = vect.transform(X_train)

len(vect.get_feature_names())
```

```python
model = svm.SVC(kernel='linear')
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
print('Accuracy',accuracy_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names())

sorted_coef_index = model.coef_[0].toarray()[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
# These reviews are now correctly identified
print(model.predict(vect.transform(['not an issue, phone is working', # in this case the model is predicting correctly 
                                    'an issue, phone is not working'])))
```

### Word Embeddings

```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.embeddings import Embedding
```

```python
docs = list(X_train)
```

```python
labels = ([y_train])
```

```python
# integer encode the documents
vocab_size = 50000
encoded_docs = [one_hot(d, vocab_size) for d in docs]
```

```python
docs[1]
```

```python
print(encoded_docs[1])
```

```python
l = list(y_train)
l[1]
```

```python
# pad documents to a max length 
max_length = max([len(s.split()) for s in docs])
print('Maximum length: %d' % max_length)
```

```python
padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
print(len(padded_docs))
print(padded_docs)
```

```python
# define the model
model = Sequential()
model.add(Embedding(vocab_size, 8, input_length=max_length))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
# summarize the model
print(model.summary())
```

```python
# fit the model
model.fit(padded_docs, labels, epochs=50, verbose=0)
```

```python
# evaluate the model
loss, accuracy = model.evaluate(padded_docs, labels, verbose=0)
print('Accuracy: %f' % (accuracy*100))
```
