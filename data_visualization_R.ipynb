```r
#This is an [R Markdown](http://rmarkdown.rstudio.com) Notebook. When you execute code within the notebook, the results appear beneath the code. 
#Try executing this chunk by clicking the *Run* button within the chunk or by placing your cursor inside it and pressing *Cmd+Shift+Enter*. 
#Installing ggplot2 package
#ggplot2 is a R package (description of why you need to import packages into R, why we are using ggplot, etc.)
library(ggplot2)
```

```r
#Now that we have imported ggplot into our R environment, we import our data that we downloaded from Kaggle. We want to keep the names of the variables as they are listed in the file, which is why we set header=TRUE.
#First, we will download the income data and then import it into R. 
#Give the file input a variable name
 income <- read.csv("../input/inc_occ_gender.csv")
```

```r
#Explore the data
#First, let's explore what types of information we have in our dataset.
colnames(income)
head(income) #shows the first couple of lines
```

```r
#For each occupation, we have information on total number of workers ("All_workers"), the number of female workers in that occupation ("F_workers"), the number of male workers ("M_workers"), and the weekly salaries for males and females ("M_weekly", "F_weekly").
#Let's start by exploring male salaries and female salaries in general, for all occupations. Since we have so much data to plot, let's do a scatterplot. 
#Just grab major category of occupations, feed it into a new data frame called subsetData

subsetData = income[grepl("^[A-Z ]+$",income$Occupation),,]
#Drop the total observation (first row)
subsetData=subsetData[-c(1),]
#Create a mini data frame called "test" with just the male and female salaries
test <- rbind(subsetData$M_weekly, subsetData$F_weekly)
#Plot "test" data frame in a bar chart
barplot(test,
        main = "Income by Gender", #title
        xlab = "Occupation", # x-axis
        ylab = "Weekly Income", #y-axis
        col=c("light blue","pink"),
        legend=c("Male","Female"), # (x, y)
        beside=TRUE)
```
